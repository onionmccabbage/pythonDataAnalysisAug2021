## Creating Charts

```python
# import the usual
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in some data
df = pd.read_excel(io='./Data/plotting_data.xls', sheet_name='anscombe')
df
I = df['dataset'] == 'I'
I
# convert the results to a DataFrame
df_I = df[I]
df_I
# plot just the 'I' values
plt.plot(df_I['x'], df_I['y'], '--x') # 'o' 'x' '--x' etc.
```

```python
df_I
```

```python
# we can plot all the data on one figure (four charts in one place)
df_I = df[df['dataset'] == 'I']
df_II = df[df['dataset'] == 'II']
df_III = df[df['dataset'] == 'III']
df_IV = df[df['dataset'] == 'IV']

fig = plt.figure(figsize=(18,6)) # the width by height
# we add subplots to this figure
ch1 = fig.add_subplot(2,2,1)
ch2 = fig.add_subplot(2,2,2)
ch3 = fig.add_subplot(2,2,3)
ch4 = fig.add_subplot(2,2,4)
ch1.plot(df_I['x'],df_I['y'], 'o')
ch2.plot(df_II['x'],df_II['y'], 'x')
ch3.plot(df_III['x'],df_III['y'], '--')
ch4.plot(df_IV['x'],df_IV['y'], '--x')
```
