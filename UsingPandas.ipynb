```python
# we need to import hte libraries for Data anlysis!
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # this convention
```

```python
# read in data from shets in an Excel workbook
googl = pd.read_excel('./data/market_data.xls', sheet_name='GOOGL', index_col='Date', parse_dates=True)
ibm = pd.read_excel('./data/market_data.xls', sheet_name='IBM', index_col='Date', parse_dates=True)
msft = pd.read_excel('./data/market_data.xls', sheet_name='MSFT', index_col='Date', parse_dates=True)
```

```python
ibm.loc['2010 march'] # or -3 or -Mar it works it out!!!
msft.loc['2013 March': '2013 May']
# mini challenge - show ALL IBM values between 2013 march 1st and 2013 April 3
ibm.loc['2013 March 1': '2013 April 13']
```

```python
# CAREFUL - changes do not persist
ibm['High']-ibm['Low']
ibm
# SO to persist any calclations, store them somewhere
difference = pd.DataFrame() # here we have a fresh empty DataFrame
difference['High-Low'] = ibm['High']-ibm['Low'] # we put a new column in our DataFram
difference
```

```python
# We can examine the structire of our DataFrames
ibm.shape
googl.size
msft.describe()
msft_1 = msft.transpose()
msft_1
```

## Filtering Data

```python
# good idea to create useful constants
TWENTY_THOU = 20*1000 #*1000
# we can generate new columns in our DataFrame
difference['GT_2K'] = difference['High-Low'] > TWENTY_THOU

difference
```

```python
# Mini challence: Make a NEW DataFrame, then populate it with
# a column for GOOGL High-Low values (calculated)
# a column for IBM High-Low values
# a column for MSFT High-Low values
three_stocks = pd.DataFrame()
three_stocks['GOOGL High Low Spread'] = googl['High']-googl['Low']
three_stocks['IBM High Low Spread'] = ibm['High']-ibm['Low']
three_stocks['Microsoft High Low Spread'] = msft['High']-msft['Low']
three_stocks # which has highest MAX high low spread?
three_stocks.describe()
```
