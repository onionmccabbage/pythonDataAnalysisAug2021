```python
# we can use Data Analysis tools with Python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# we can read in data from excel (or csv or web ...)
my_data = pd.read_excel(io='./data/sample_data.xls', sheet_name='SP500', index_col='Symbol')
my_data # this is called a 'DataFrame'
# my_data.describe()
my_data.tail(12)
```

```python
# we can access rows and columns
my_data[['Name', 'Sector', 'Price']] # CAREFUL-we need TWO square brackets
```

```python
# access rows using 'loc' (for location)
my_data.loc[['MMM', 'C', 'GOOGL']]
```

```python
# we often make a collection of the rows or columns we want
cols = ['Name', 'Sector', 'Price'] # this is a list
rows = ['IBM', 'MMM', 'C']
my_data[cols]
my_data.loc[rows]
```

```python
# we can analyse rows or columns
my_data['Earnings/Share'].median() # or mean, max, min, count etc
```

```python
#We can use Numpy to analyse columns of data
arr1 = np.array ([1,2,3,4,5])
arr2 = np.array([10,20,30,40,50])
arr1, arr2
# Fast mathematics on arrays
arr1**arr2

arr1 < arr2
```

```python
# we use Numpy to operate VEERY FAST on Pandas DataFrames
my_data['52 Week High'] - my_data['52 Week Low']
my_data['Earnings/Share']<5
```
